```python
import numpy as np
file = "./data/glove.twitter.27B.100d.txt"
wordvecs = {}
if file is not None:
    print("Reading glove file : ", file)
    found = 0
    with open(file, "r") as f:
        for line in f:
            temp = line.split()
            word = temp[0].rstrip(">").lstrip("<")
            vecs = [float(k) for k in temp[1:]][:100]
            wordvecs[word] = np.array(vecs)
            found += 1
```

```python
np.dot(wordvecs["fucc"], wordvecs["run"])
```
